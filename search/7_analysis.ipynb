```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv("data/master20250929.csv")
df = pd.read_csv("data/master20260210analysis.csv") # revision
df["year"] = df["year"].astype(int)
```

```python
country_counts = (
    df.groupby("country")["id"]
      .count()
      .rename("deployments")
      .reset_index()
      .sort_values("deployments", ascending=False))
country_counts
```

```python
# count deployments per year
year_counts = (
    df.groupby("year")["id"]
      .count()
      .rename("deployments")
      .reset_index()
      .sort_values("year"))
year_counts
```

```python
mode_counts = (
    df.groupby(["year", "mode"])["id"]
      .count()
      .unstack(fill_value=0)
      .sort_index())
mode_counts
```

```python
stack_order = ["road", "rail", "air", "waterborne", "cross_modal"]
mode_counts_ordered = mode_counts[stack_order]

colors = [
    "#5D877D",  # Road
    "#4F3857",  # Rail
    "#B05653",  # Air
    "#FFEABA",  # Waterborne
    "#E8A680"   # Cross modal
]

ax = mode_counts_ordered.plot(
    kind="bar",
    stacked=True,
    figsize=(10, 6),
    color=colors
)

ax.set_xlabel("Year")
ax.set_ylabel("Number of AI Deployments")

legend_labels = ["Road", "Rail", "Air", "Waterborne", "Cross modal"]
ax.legend(
    legend_labels,
    title="Mode",
    loc="upper left",
    frameon=False
)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

#plt.savefig("pix/mode_counts.tiff", format="tiff", dpi=300)
plt.tight_layout()
plt.show()
```

```python
uc_counts = (
    df.groupby("category")
      .agg(
          n=("category", "size"),
          cities=("city", lambda x: ", ".join(
              sorted(
                  f"{city} ({country})"
                  for city, country in zip(
                      df.loc[x.index, "city"],
                      df.loc[x.index, "country"]
                  )
              )
          ))
      )
      .sort_values("n", ascending=False)
      .reset_index()
)

uc_counts
```
